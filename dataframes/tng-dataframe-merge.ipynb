```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('./dataframe_all_tng.csv')
```

```python
imdb_df = pd.read_csv('./tng_imdb_data.csv')
```

```python
imdb_df.rename(columns={'ep_name': 'ep_title'}, inplace=True)
```

```python
df.head()
```

```python
imdb_df.head()
```

```python
df['ep_title_formatted'] = [''.join(e for e in title if e.isalnum()).lower() for title in df['ep_title']]
```

```python
df.head()
```

```python
imdb_df['ep_title_formatted'] = [''.join(e for e in title if e.isalnum()).lower() for title in imdb_df['ep_title']]
```

```python
imdb_df.head()
```

```python
imdb_set = set(imdb_df['ep_title_formatted'])
df_set = set(df['ep_title_formatted'])
```

```python
imdb_replace = []
for ep in imdb_set:
    if ep not in df_set:
        imdb_replace.append(ep)
        print(ep)
```

```python
df_replace = []
for ep in df_set:
    if ep not in imdb_set:
        df_replace.append(ep)
        print(ep)
```

```python
df_replace.sort()
```

```python
df_replace
```

```python
imdb_replace.sort()
```

```python
imdb_replace
```

```python
replace_master = list(zip(df_replace, imdb_replace))
```

```python
replace_master_df = pd.DataFrame(replace_master, columns=['df', 'imdb'])
```

```python
len(replace_master_df)
```

```python
replace_master_df.iloc[0,0]
```

```python
replace_master_df.iloc[0,1]
```

```python
for i in list(range(0,19)):
    df.replace(to_replace=replace_master_df.iloc[i,0], value=replace_master_df.iloc[i,1], inplace=True)
```

```python
df
```

```python
merged_df = pd.merge(df, imdb_df, on='ep_title_formatted', how='outer')
```

```python
merged_df.head()
```

```python
merged_df.drop('ep_title_x', axis=1, inplace=True)
```

```python
merged_df.to_csv('merged_tng.csv', index=False)
```
